#### Importing required libraries

```python
from statsbombpy import sb
import pandas as pd
import os
```

#### View available competitions in the dataset

```python
comps = sb.competitions()
print(comps)
```

```python
# Get list of all competitions
comp_dict = {}
for competition in comps["competition_name"]:
    if competition not in comp_dict:
        comp_dict[competition] = True

competition_list = comp_dict.keys()

print(competition_list)
```

```python
# Search competittion ID for 2022 World Cup
for i in range(len(comps)):
    if comps["competition_name"][i] == "FIFA World Cup" and comps["season_name"][i] == "2022":
        print("2022 World Cup Competition ID: ", comps["competition_id"][i])
        print("2022 World Cup Season Id: ", comps["season_id"][i])
```

#### Extract data for 2022 FIFA World Cup matches

```python
competition_id = 43
season_id = 106

matches = sb.matches(competition_id=competition_id, season_id=season_id)

print(matches)
```

```python
# Extracting data from 2022 FIFA World Cup matches

data_file = "../data/wc_2022_events.csv"

match_ids = matches['match_id'].unique()
all_events = []

for m_id in match_ids:
    match_events = sb.events(match_id=m_id)
    all_events.append(match_events)

events_df = pd.concat(all_events, ignore_index=True)
```

```python
# Caching the data in a csv file

if not os.path.exists(data_file):
    events_df.to_csv(data_file, index=False)
```
